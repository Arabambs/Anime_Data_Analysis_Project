```python
Q1 = df["Members"].quantile(0.25)
Q3 = df["Members"].quantile(0.75)


IQR = Q3 - Q1

upper = np.where(df["Members"] >= (Q3+1.5*IQR))

print("upper bound", upper)

lower = np.where(df["Members"] <= (Q1-1.5*IQR))

print("lower bound", lower)
```

```python
from scipy import stats
from scipy.stats import zscore

z_scores = stats.zscore(df['Members'])

abs_z_scores = np.abs(z_scores)

filtered_entries = (abs_z_scores < 3).all(axis=2)

new_df = df[filtered_entries]


print(new_df)
```
